```python
# deconvolutional layer

# deconvolutional layer has input and output channels a kernel size and if we want an activation layer or not
def deconvolutional_layer(input_channels, output_channels, kernel_size = 3, activation_layer = True):
    # Nearest neighbour upsampling with a scale factor of 2 and then we apply a convolutional layer onto the upsampling
    layers = [nn.UpsamplingNearest2d(scale_factor = 2),
             nn.Conv2d(input_channels, output_channels, stride = 1, kernel_size = kernel_size, padding = kernel_size // 2)]
    if activation_layer:
        layers.append(nn.ReLU())
    return nn.Sequential(*layers)
```

```python
# Takes our model a loss function and the validation set dataloader
def eval(model, loss_func, valid_dl, epoch=0):
    # whenever you are using dropout and batch normalisation these are only done during training and not during inference
    # so we use model.eval() whenever we want to toggle them on and off
    model.eval()
    # use with torch.no_grad()
    with torch.no_grad():
        tot_loss,count = 0.,0
        for xb,_ in valid_dl:
            pred = model(xb)
            n = len(xb)
            count += n
            tot_loss += loss_func(pred,xb).item()*n
    print(epoch, f'{tot_loss/count:.3f}')
```

```python
# Whenever we are fitting the loss function compares the model prediction image with the value it was input itself
def fit(epochs, model, loss_func, opt, train_dl, valid_dl):
    for epoch in range(epochs):
        model.train()
        for xb,_ in train_dl:
            loss = loss_func(model(xb), xb)
            loss.backward()
            opt.step()
            opt.zero_grad()
        eval(model, loss_func, valid_dl, epoch)
     
```
